```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
fake_news_df = pd.read_csv('/kaggle/input/fakenewscorpusoriginal/news-data-12p.csv')

# Counting NaN per column
missing_values_count = fake_news_df.isnull().sum()

# Plotting
plt.figure(figsize=(10, 6))
missing_values_count.plot(kind='bar', color='skyblue')
plt.title('Blank value for Each Column')
plt.xlabel('Columns')
plt.ylabel('Missing Value Count')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

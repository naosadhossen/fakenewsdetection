```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('/content/drive/MyDrive/LUT/ISS/labeled-new-strictness-high.csv')

# Group the data by domain and label, and count the occurrences
domain_label_counts = df.groupby(['domain', 'label']).size().reset_index(name='count')

# Pivot the data to have labels as columns
pivot_table = domain_label_counts.pivot(index='domain', columns='label', values='count').fillna(0)

# Count domains with label 0, label 1, and both labels
count_label_0 = (pivot_table[0] > 0).sum()
count_label_1 = (pivot_table[1] > 0).sum()
count_both_labels = ((pivot_table[0] > 0) & (pivot_table[1] > 0)).sum()

# Create a bar plot
labels = ['Label 0', 'Label 1', 'Both Labels']
counts = [count_label_0, count_label_1, count_both_labels]

plt.bar(labels, counts, color=['blue', 'green', 'orange'])
plt.xlabel('Label')
plt.ylabel('Number of Domains')
plt.title('Counts of Domains by Label')
plt.show()
```
